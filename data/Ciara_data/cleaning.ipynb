```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('average-length-of-stay.csv')

# Display first few rows to verify
print("Original data shape:", df.shape)
print(df.head())
```

```python
# For each country (Entity), keep only the row with the maximum Year
df_filtered = df.loc[df.groupby('Entity')['Year'].idxmax()]

# Sort by Entity 
df_filtered = df_filtered.sort_values('Entity').reset_index(drop=True)

print("Filtered data shape:", df_filtered.shape)
print(df_filtered.head(10))
```

```python
# Save to xlsx file
df_filtered.to_excel('average-length-of-stay-filtered.xlsx', index=False)
print("File saved successfully as 'average-length-of-stay-filtered.xlsx'")
```
